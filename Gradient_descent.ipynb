æœ¬ç­†è¨˜æœ¬ç¤ºç¯„å¦‚ä½•ä½¿ç”¨æ¢¯åº¦ä¸‹é™æ³• (Gradient Descent) ä¾†æœ€å°åŒ–ç°¡å–®å‡½æ•¸ f(x) = xÂ²ã€‚

```
import numpy as np
import matplotlib.pyplot as plt
```

## å®šç¾©ç›®æ¨™å‡½æ•¸èˆ‡å°æ•¸

```
# ç›®æ¨™å‡½æ•¸ f(x) = x^2
def f(x):
    return x**2

# å°æ•¸ f'(x) = 2x
def df(x):
    return 2*x
```

## å¯¦ä½œæ¢¯åº¦ä¸‹é™

```
# åˆå§‹å€¼è¨­å®š
x = 5.0          # èµ·é»
lr = 0.1         # å­¸ç¿’ç‡
epochs = 25      # è¿­ä»£æ¬¡æ•¸

x_history = [x]

for i in range(epochs):
    grad = df(x)
    x = x - lr * grad
    x_history.append(x)

print(f"æœ€çµ‚çµæœ x = {x:.5f}, f(x) = {f(x):.5f}")
```

## è¦–è¦ºåŒ–æ¢¯åº¦ä¸‹é™éç¨‹

```
# ç¹ªè£½å‡½æ•¸èˆ‡ä¸‹é™è»Œè·¡
x_vals = np.linspace(-6, 6, 200)
y_vals = f(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label="f(x) = xÂ²")
plt.scatter(x_history, [f(x) for x in x_history], color='red', label="Gradient Descent Steps")
plt.title("Gradient Descent Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

## æ”¶æ–‚éç¨‹åœ–

```
plt.plot(range(len(x_history)), [f(x) for x in x_history], marker='o')
plt.title("Loss Convergence over Iterations")
plt.xlabel("Iteration")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
